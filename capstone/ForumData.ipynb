```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
def getThread(thread_url):
    #print('+++++++++++++++++++++++++++++++' + thread_url)
    one_thread = ""
    r_thread = requests.get(thread_url)
    thread_soup = BeautifulSoup(r_thread.content)
    thread_items = thread_soup.find_all('div', {'class': 'bbWrapper'})
    for item in thread_items:
        one_thread = one_thread + item.get_text()
    return one_thread
    #print(thread_items)
    #print('------------------------------' + thread_url)
    
```

```python
main_url = 'https://www.diabetesdaily.com'
forum_url = '/forum/forums/type-2-diabetes.14/'

#create directory in file system to store all the threads for a given forum
main_dir = './data/www.diabetesdaily.com'
forum_dir = main_dir + forum_url

if os.path.exists(main_dir) != True :
    os.makedirs(main_dir)
if os.path.exists(forum_dir) != True:
    os.makedirs(forum_dir)

max_pages = 1847
#max_pages = 3

for i in range( max_pages+1):

    page_url_part = ''
    if (i+1) != 1:
        page_url_part = 'page-' + str(i+1)
    #print(main_url+ forum_url + page_url_part)
    r = requests.get(main_url+ forum_url + page_url_part)
    
    soup = BeautifulSoup(r.content)

    items = soup.find_all('div', {'class': 'structItem-title'})
    print('+++++++++++++++++++++ PAGE ' + str(i))
    for item in items:
        thread_url = item.find('a')['href']
        item_title = item.find('a').contents[0]        
        #print('Thread URL: ' + thread_url + ' item_title : ' + item_title)
        thread = getThread(main_url+thread_url)
        #item_title = item_title.replace('!', ' ')
        item_title = item_title.replace('/', ' ')
        #print(forum_dir+ '\"' + item_title + '\"')
        
        #Write thread out to the file with title as the file name
        fo = open(forum_dir+item_title, 'w')
        fo.write(thread)
        fo.close()
        
        
    print('--------------------- PAGE ' + str(i))
```
